## Energía promedio para señales EEG:

Creación de la función para calcular la energía promedio de cada canal:

```python
# Se crea una funcion para calcular la energia
def energiaPromedio(signal):
  canales, puntos, epocas = signal.shape

  # Se calcula la energia de cada epoca
  energia_epocas = np.sum(signal**2, axis=1) #suma sobre las columnas, a lo largo de los puntos en el tiempo para cada fila (canal)
  # Se calcula la energia promedio de cada canal
  energia_promedio = np.mean(energia_epocas, axis=1)

  return energia_promedio
```

# Sujetos Control

Creación de dataframe para energía promedio por época para cada sujeto:

```python
energiasControl = []
for archivo in archivos_control:
    sujetos_control = sio.loadmat(extracted_control + "/control/" + archivo)

    # Extraer los datos del archivo .mat
    data_control = sujetos_control['data']
    energia_promedio = energiaPromedio(data_control)
    energiasControl.append(energia_promedio)


# Crear un DataFrame con los resultados de todos los sujetos
df_control = pd.DataFrame(energiasControl, columns = [f'canal_{i+1}' for i in range(energia_promedio.shape[0])])
df_control['sujeto'] = archivos_control
df_control.set_index('sujeto', inplace=True)

# Guardar los resultados en un archivo CSV
df_control.to_csv('energia_promedio_control_sujetos.csv', index=False)

# Imprimir el DataFrame para ver los resultados
df_control
```

# Sujetos Parkinson

Creación de dataframe para energía promedio por época para cada sujeto:

```python
energiasParkinson = []
for archivo in archivos_parkinson:
    sujetos_parkinson = sio.loadmat(extracted_parkinson + "/parkinson/" + archivo)

    # Extraer los datos del archivo .mat
    data_parkinson = sujetos_parkinson['data']
    energia_promedio = energiaPromedio(data_parkinson)
    energiasParkinson.append(energia_promedio)

# Crear un DataFrame con los resultados de todos los sujetos
df_parkinson = pd.DataFrame(energiasParkinson, columns = [f'canal_{i+1}' for i in range(energia_promedio.shape[0])])
df_parkinson['sujeto'] = archivos_parkinson
df_parkinson.set_index('sujeto', inplace=True)

# Guardar los resultados en un archivo CSV
df_parkinson.to_csv('energia_promedio_parkinson_sujetos.csv', index=False)

# Imprimir el DataFrame para ver los resultados
df_parkinson
```

# Comparación entre canales de cada grupo poblacional:

**Normalidad:**

Prueba de Kolmogorov-Smirnov

**Hipótesis:**

p > alpha

*   H0: Los datos parecen normales (No se puede rechazar H0)
*   H1: Los datos no parecen normales (Se rechaza H0)

**Homocedasticidad:**

Prueba de Levene

**Hipótesis:**

*   H0: Los grupos tienen varianzas iguales
*   H1: Los grupos tienen varianzas diferentes

**Independencia**

Se asume que los grupos son independientes

**U Mann-Whitney-Wilcoxon**

* HO: No hay diferencia (en términos de tendencia central) entre los dos grupos de la población.
* H1: Existe una diferencia (con respecto a la tendencia central) entre los dos grupos de la población.

Las condiciones del u-test son:

* Los datos tienen que ser independientes.

* Los datos tienen que ser ordinales.

* No es necesario asumir que las muestras se distribuyen de forma normal.

* Igualdad de varianza entre grupos (homocedasticidad).

```python
resultados = []

for canal in df_control.columns:
  energia_promedio_control = df_control[canal].values
  energia_promedio_parkinson = df_parkinson[canal].values

  # Realizar la prueba de normalidad
  normalidad_control = stats.kstest(energia_promedio_control, 'norm')
  normalidad_parkinson = stats.kstest(energia_promedio_parkinson, 'norm')
  # Realizar prueba de homocedasticidad
  homocedasticidad = stats.levene(energia_promedio_control, energia_promedio_parkinson)
  if normalidad_control.pvalue > 0.05 and normalidad_parkinson.pvalue > 0.05 and homocedasticidad.pvalue > 0.05:
    t_statistic, p_value = stats.ttest_ind(energia_promedio_control, energia_promedio_parkinson)
    prueba = 't-test'
  else:
    t_statistic, p_value = stats.mannwhitneyu(energia_promedio_control, energia_promedio_parkinson)
    prueba = 'mann-whitney'

  resultados.append((canal, prueba, p_value, normalidad_control.pvalue, normalidad_parkinson.pvalue, homocedasticidad.pvalue))

df_resultados = pd.DataFrame(resultados, columns=['Canal', 'Prueba','P-value (prueba)', 'Normalidad Control', 'Normalidad Parkinson', 'Homocedasticidad'])
df_resultados
```

## Análisis de Resultados:

Tras aplicar la prueba de homocedasticidad a los canales entre los distintos grupos poblacionales, se concluye que estos presentan homocedasticidad. Esto implica que no hay suficiente evidencia para rechazar la hipótesis nula, lo que sugiere que las varianzas entre los grupos son aproximadamente iguales. Sin embargo, al detectarse una violación del supuesto de normalidad, se hace necesario realizar análisis por técnicas no paramétricas. Por lo tanto se procede a realizar la prueba de Mann-Whitney como siguiente paso en el análisis.

En cuanto a los resultados de la prueba U de Mann-Whitney, si no se rechaza la hipótesis nula, esto indica que no hay suficiente evidencia estadística para afirmar la existencia de diferencias significativas entre las distribuciones de los dos grupos comparados. En el contexto de la evaluación de la señal de los canales en pacientes con Parkinson y en pacientes sanos, si no se rechaza la hipótesis nula, se puede inferir que los datos no muestran la presencia de una diferencia significativa en la señal del canal entre ambos grupos. Sin embargo, es importante mencionar que esto no descarta la posibilidad de una diferencia real, ya que otras variables o limitaciones en los datos podrían estar influyendo en la ausencia de diferencias significativas observadas. Por ejemplo, un tamaño de muestra pequeño puede limitar la capacidad de detectar diferencias estadísticamente significativas, lo que reduce el poder estadístico de la prueba, otra posible causa podría deberse a que las señales EEG son contienen ruido y pueden variar entre individuos debido a factores como el estado emocional, el cansancio o la actividad cerebral no relacionada, que aunque con el análisis realizado no puede afirmarse, se encuentra entre las posibilidades de este análisis.
