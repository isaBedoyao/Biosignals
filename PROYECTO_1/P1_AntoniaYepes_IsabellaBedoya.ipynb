# Enfermedad de Parkinson

La enfermedad de Parkinson es un trastorno del cerebro que provoca movimientos involuntarios o incontrolables, como temblores, rigidez y dificultad con el equilibrio y la coordinación [2].

### Diagnóstico [3]:

La enfermedad de Parkinson (EP) es un diagnóstico “clínico”. Esto significa que se utilizan los antecedentes, los síntomas y el examen físico de la persona para realizar el diagnóstico. No existe una prueba de laboratorio o de diagnóstico por imágenes específica que pueda diagnosticar la EP. Sin embargo, se pueden utilizar ciertas pruebas, como la resonancia magnética del cerebro (MRI cerebral), una exploración del transportador de dopamina (DaT scan) o un análisis de sangre para respaldar el diagnóstico de EP o para descartar otras afecciones médicas que puedan imitar la EP.

Los criterios más nuevos para diagnosticar la enfermedad de Parkinson fueron desarrollados por la Sociedad Internacional de Parkinson y Trastornos del Movimiento (MDS) y reflejan la comprensión más actual de la EP.

Para considerar un diagnóstico de enfermedad de Parkinson, una persona debe tener bradicinesia (lentitud de movimiento). Además de bradicinesia, una persona también debe tener uno o más de los siguientes síntomas [5]:

* Temblor o sacudida en una extremidad que ocurre mientras está en reposo.
* Rigidez o rigidez de los brazos, piernas o tronco.
* Problemas de equilibrio y caídas.

### Reto del diagnóstico temprano [4]:

* **Falta de biomarcadores específicos y sensibles:**
Actualmente, no existen biomarcadores confiables que permitan identificar con certeza las fases prediagnósticas de la EP. Esto es particularmente importante para poder intervenir de manera temprana antes de que la enfermedad se manifieste clínicamente​.

* **Dificultad para diferenciar la EP de otros trastornos:**
En las primeras etapas de la EP, es difícil diferenciarla de otras condiciones parkinsonianas atípicas, como la atrofia multisistémica o la parálisis supranuclear progresiva, ya que estas condiciones comparten algunos síntomas motores iniciales​.

* **Subestimación de los síntomas prodrómicos:**
Los síntomas prodrómicos no motores, como la hiposmia (pérdida del olfato), el estreñimiento o los trastornos del sueño REM, pueden presentarse muchos años antes de los síntomas motores, pero son inespecíficos y no siempre se reconocen como señales tempranas de la EP​.

* **Variabilidad en la progresión de la enfermedad:**
La EP es una enfermedad heterogénea con diferentes subtipos clínicos, lo que hace que la progresión de la enfermedad varíe de un paciente a otro. Esta variabilidad complica el desarrollo de estrategias de diagnóstico temprano, ya que no todos los pacientes siguen el mismo patrón​.

* **Imprecisión en la caracterización clínica temprana:**
Incluso con los avances en los criterios clínicos y la validación de nuevas pruebas, la precisión diagnóstica en las etapas tempranas sigue siendo limitada. En algunos estudios, hasta un 10% de los pacientes diagnosticados clínicamente con EP presentaban patologías alternativas en el examen postmortem​.

# Enfermedad de Huntington

La enfermedad de Huntington es un trastorno hereditario y degenerativo que avanza con el tiempo, causando la destrucción de neuronas en ciertas regiones del cerebro. Esta degradación neuronal lleva a la aparición de movimientos involuntarios, deterioro cognitivo y problemas emocionales en quienes la padecen [6].

### Diagnostico [7]:
El diagnóstico de la enfermedad de Huntington se basa en un examen físico general, preguntas sobre la historia médica familiar, y una evaluación neurológica y de salud mental.

* **Examen neurológico:**
Un neurólogo evalúa síntomas motores (reflejos, fuerza muscular, equilibrio), síntomas sensoriales (tacto, visión, audición) y síntomas psiquiátricos (estado de ánimo, estado mental).

* **Pruebas neuropsicológicas:**
Se realizan pruebas para medir la memoria, razonamiento, agilidad mental, habilidades lingüísticas y razonamiento espacial. Si es necesario, un psicólogo puede hacer pruebas más completas.

* **Evaluación de salud mental:**
Un psiquiatra evalúa el estado emocional, patrones de comportamiento, juicio, habilidades de afrontamiento, signos de pensamiento desordenado y evidencia de abuso de sustancias.

* **Imágenes cerebrales:**
Pruebas como la resonancia magnética (MRI) o la tomografía computarizada (CT) muestran imágenes detalladas del cerebro. Estas imágenes pueden revelar cambios en áreas afectadas por la enfermedad, aunque no siempre son visibles en las etapas tempranas.

* **pruebas genéticas:**
Si se sospecha la enfermedad de Huntington, puede recomendarse una prueba genética para confirmar el diagnóstico, especialmente si no hay antecedentes familiares claros. Un consejero genético explicará los beneficios y riesgos de realizar la prueba.

* **Prueba genética predictiva:**
Se puede realizar una prueba genética predictiva si hay antecedentes familiares de la enfermedad pero aún no hay síntomas. No indica cuándo comenzarán los síntomas ni cuáles aparecerán primero. Se hace después de consultar con un consejero genético y puede implicar riesgos relacionados con el seguro o el empleo.

### Reto del diagnostico temprano:

* **Síntomas tempranos inespecíficos:**
Los síntomas iniciales suelen incluir alteraciones psiquiátricas (depresión, irritabilidad) y problemas cognitivos (dificultades en la planificación y memoria), que son comunes en muchas otras enfermedades, lo que puede llevar a diagnósticos erróneos o retrasados hasta que aparecen los síntomas motores típicos como la corea [8].

* **Variabilidad en la aparición de los síntomas:**
La edad de inicio y la progresión de los síntomas varían considerablemente, incluso dentro de la misma familia. Esto dificulta la predicción precisa de cuándo aparecerán los síntomas, lo que representa un reto para el diagnóstico temprano [9].

* **Limitaciones de las pruebas genéticas:**
Aunque las pruebas genéticas pueden confirmar la presencia de la mutación en el gen HTT, no permiten predecir con exactitud cuándo comenzarán los síntomas ni su severidad. Esto genera incertidumbre en los portadores de la mutación[10].

* **Dilemas éticos y emocionales:**
Las pruebas predictivas pueden realizarse en personas asintomáticas con antecedentes familiares, pero saber que se es portador de la mutación sin tener tratamiento para prevenir o modificar la enfermedad puede generar una carga emocional significativa [11].

# Esclerosis Lateral Amiotrófica (ELA)

Se trata de una enfermedad neurológica que afecta las neuronas motoras que controlan el movimiento muscular voluntario. Es una enfermedad progresiva, lo que significa que los síntomas empeoran con el tiempo [12].

### Diagnóstico [12]:

No hay una única prueba capaz de confirmar con certeza la esclerosis lateral amiotrófica (ELA). El proceso de diagnóstico involucra un examen físico y una revisión del historial médico del paciente. Se realiza un examen neurológico para evaluar reflejos, fuerza muscular y otras respuestas, y se repite periódicamente para monitorizar la progresión de síntomas como debilidad y atrofia muscular, así como espasticidad.

Para descartar otras enfermedades y confirmar el diagnóstico, se utilizan las siguientes pruebas:

- **Electromiografía (EMG):** Esta técnica mide la actividad eléctrica en las fibras musculares y puede ser útil para diagnosticar la ELA.
- **Estudio de conducción nerviosa:** Evalúa la actividad eléctrica en los nervios y músculos, verificando la capacidad del nervio para transmitir señales al músculo.
- **Resonancia magnética (MRI):** Utiliza un campo magnético y ondas de radio para crear imágenes detalladas del cerebro y la médula espinal sin necesidad de cirugía.
- **Análisis de sangre y orina:** Según los síntomas y los resultados de otras pruebas, estos análisis pueden ser solicitados para descartar otras posibles enfermedades.
- **Biopsia muscular:** Si se sospecha una enfermedad muscular diferente a la ELA, se puede realizar una biopsia muscular bajo anestesia local, extrayendo una pequeña muestra para su análisis en laboratorio.

###Reto del diagnóstico temprano:

El diagnóstico de la Esclerosis Lateral Amiotrófica (ELA) es complejo debido a la falta de una prueba específica. Se basa principalmente en la exclusión de otras enfermedades que pueden presentar síntomas similares, como la neuropatía motora multifocal. La heterogeneidad de los síntomas, que pueden variar entre manifestaciones bulbares y espinales, añade dificultad al proceso. En promedio, el diagnóstico se retrasa entre 9 y 15 meses desde el inicio de los síntomas, lo que afecta la intervención temprana. Aunque existen criterios clínicos revisados, como los del Awaji–Shima, estos no siempre capturan aspectos no motores de la enfermedad, lo que puede complicar aún más el diagnóstico [13]


# Detrend scipy

Los datos de series temporales son una colección de observaciones realizadas a lo largo del tiempo. Estas observaciones pueden mostrar una tendencia.

En términos sencillos, una tendencia significa que los valores de la serie temporal pueden aumentar o disminuir con el tiempo.

A veces se requiere analizar los datos sin la tendencia para capturar otros patrones como la estacionalidad o el ruido. En estas situaciones, la eliminación de tendencias puede ser de ayuda.[14]

Se supone que la serie temporal es una combinación aditiva de una tendencia y otros componentes.
$$
Y(t) = T(t) + S(t) + e(t)
$$

Donde:

- Y(t) es la serie temporal original.
- T(t) es la tendencia, que muestra el comportamiento a largo plazo.
- S(t) es el componente estacional, que representa los patrones repetitivos o cíclicos.
- e(t) es el ruido, que captura las fluctuaciones aleatorias o no explicadas.

Entonces, para eliminar la tendencia de una serie temporal, se necesita estimar el componente de tendencia y luego restarlo de los datos originales, dejándolo solo con el componente estacional y el ruido. Para esto existen 2 tipos, lineal o constante.

1. **Detrending con un Modelo Constante (Scipy)** [15]

Cuando se usa el modelo constante, se elimina la media de la serie temporal. Matemáticamente, esto se puede expresar como :


$$
\text{Detrended}(x(t)) = x(t) - \mu
$$

Donde:

- x(t) es el valor de la serie temporal en el tiempo \( t \).
- mu es la media de los valores de la serie temporal.

Eliminando la media, lo que se obtiene es una serie centrada alrededor de cero. Esto es útil si la serie tiene una tendencia horizontal (sin pendiente).

2. **Detrending con un Modelo Lineal (Scipy)** [16]

Con un modelo lineal, se ajusta una línea recta a la serie temporal y se resta de los datos. Matemáticamente, este ajuste es:
La tendencia lineal ajustada se puede describir mediante la siguiente fórmula:

$$
y(t) = a*t + b
$$

Donde:

- y(t) es la tendencia lineal ajustada,
- a es la pendiente de la línea,
- b es la intersección con el eje \( y \).

El objetivo es encontrar los valores de \( a \) y \( b \) que minimicen la diferencia entre los puntos de los datos \( x(t) \) y la línea ajustada \( y(t) \). Esto se logra minimizando la suma de los errores al cuadrado, que se expresa como:

$$
\text{Minimizar} \sum_{t=1}^{n} \left( x(t) - (a*t + b) \right)^2
$$

Aquí:

- x(t) son los datos observados,
- (a*t + b) es el valor predicho por la línea ajustada,
- El término \( x(t) - (a*t + b) \)^2 es el error al cuadrado entre el valor observado y el valor predicho.

Cuando se llama a `detrend(data, type='linear')`, **scipy** realiza este ajuste por mínimos cuadrados internamente:

1. Ajusta una línea recta \( y(t) = a*t + b \) a los datos de entrada.
2. Resta esta línea ajustada de los datos originales, produciendo una versión "detrended" de la serie temporal, es decir:

$$
\text{Detrended}(x(t)) = x(t) - (a \cdot t + b)
$$


# Filtro Hampel

El filtro de Hampel es un método robusto para detectar y manejar valores atípicos en datos de series temporales. Se basa en la desviación absoluta media (MAD) y emplea una ventana móvil para la identificación de valores atípicos. MAD es una medida robusta de dispersión de datos, calculada como la mediana de las desviaciones absolutas con respecto al valor medio [17].

La configuración del filtro Hampel implica dos parámetros:

Tamaño de la ventana : este parámetro determina el tamaño de la ventana móvil que se utiliza para evaluar cada punto de datos. Básicamente, define el alcance dentro del cual buscamos valores atípicos.

Umbral : la selección cuidadosa del umbral es esencial para evitar que se active la detección de valores atípicos en datos valiosos.

### Proceso de Detección [16]

Para cada \( x_i \) en la serie temporal, se define una ventana \( W_i \) de tamaño \( 2k + 1 \), centrada en \( x_i \):

$$
W_i = \{x_{i-k}, \dots, x_i, \dots, x_{i+k} \}
$$

La **mediana** de los valores en la ventana \( W_i \) es:

$$
m_i = \text{Mediana}(W_i)
$$

Para medir la variabilidad en la ventana, se utiliza la desviación mediana absoluta (**MAD**):

$$
M_i = \text{Mediana}(\{|x_{i-k} - m_i|, \dots, |x_i - m_i|, \dots, |x_{i+k} - m_i|\})
$$

La **desviación estándar aproximada** \( \sigma_i \) se estima usando el factor \( 1.4826 \):

$$
\sigma_i = 1.4826 \cdot M_i
$$

### Criterio de Detección de Outliers

Un valor \( x_i \) se considera un outlier si cumple la condición:

$$
|x_i - m_i| > \kappa \cdot \sigma_i
$$

Donde \( k \) es el número de desviaciones estándar para clasificar un valor como outlier (por ejemplo, \( k = 3 \) si se utilizan 3 desviaciones estándar).

Si \( x_i \) es un outlier, se reemplaza por la mediana \( m_i \).

### Ventanas Truncadas

Para los primeros y últimos \( k \) elementos de la serie, las ventanas se ajustan para incluir solo los valores disponibles, modificando así los cálculos de \( V_i \) y \( M_i \).


# Entropía de permutación

La entropía de permutación mide el grado de desorden en una secuencia de datos y está basada en la probabilidad de que ciertas configuraciones ordenadas ocurran dentro de un sistema. Este concepto es importante en bioingeniería para analizar la complejidad y variabilidad en sistemas biológicos, como en este caso, en el análisis de la marcha humana, donde se busca capturar patrones en señales temporales de movimiento.

Matemáticamente, la entropía de permutación considera una serie temporal y transforma los valores en patrones ordinales, es decir, la posición relativa de los valores en función de su orden. Luego, se mide la probabilidad de que cada patrón ocurra dentro de la serie. La fórmula que rige la entropía de permutación es la entropía de Shannon aplicada a las probabilidades de cada patrón observado:

$$
H = -\sum_{i=1}^{n} p_i \log(p_i)
$$

donde p_i es la probabilidad de ocurrencia del patrón ordinal (i), y (n) es el número total de patrones distintos posibles.

Desde una perspectiva de bioingeniería, la entropía de permutación puede aplicarse para analizar señales como la cinemática de las articulaciones durante el movimiento. Por ejemplo, al estudiar la marcha, la entropía de permutación puede ayudar a determinar qué tan predecible o aleatorio es el patrón de movimiento de un individuo. Si la entropía es baja, el patrón es más regular, indicando que el sistema es más rígido o predecible, lo cual podría estar asociado con patologías o envejecimiento. Una entropía alta sugiere mayor desorden o variabilidad, lo cual puede reflejar un comportamiento más adaptativo o saludable.

La entropía de permutación es útil en sistemas que presentan variabilidad y que son difíciles de predecir, como en el análisis de la marcha, donde se emplea para cuantificar la complejidad de los movimientos humanos a través del tiempo [18]

Función para cargar archivo .ts, extraer señales de interés, aplicar detrend y filtro hampel

```python
def procesar_archivo_ts(ruta_archivo, K=3, Thresh=6):
    columnas= ['Elapsed Time (sec)', 'Left Stride Interval (sec)', 'Right Stride Interval (sec)', 'Left Swing Interval (sec)',
    'Right Swing Interval (sec)',	'Left Swing Interval (% of stride)', 'Right Swing Interval (% of stride)', 'Left Stance Interval (sec)', 'Right Stance Interval (sec)', 'Left Stance Interval (% of stride)',
    'Right Stance Interval (% of stride)', 'Double Support Interval (sec)',	'Double Support Interval (% of stride)']

    datos = pd.read_csv(ruta_archivo, sep="\t", names=columnas)

    señales_interes = datos[['Left Stride Interval (sec)', 'Right Stride Interval (sec)', 'Left Swing Interval (sec)',
    'Right Swing Interval (sec)']].values

    # Convertir a ndarray
    señales_ndarray = np.array(señales_interes)

     # Eliminar la tendencia de cada señal usando detrend
    señales_detrend = detrend(señales_ndarray)

    # Aplicar el filtro Hampel para eliminar los valores atípicos
    señales_filtradas = filtro_hampel(señales_detrend, K, [Thresh])

    return señales_filtradas
```

Función que reciba una señal de marcha y calcule el coeficiente de variación (CV) y la (SD)

```python
def calcular_cv_sd(signal):
    # Calcular la media de la señal
    media = np.mean(signal)

    # Calcular la desviación estándar
    sd = np.std(signal)

    # Calcular el coeficiente de variación (CV)
    cv = (sd / media) * 100

    return cv, sd
```

Crear una función que reciba una señal de marcha y permita calcular la entropía de permutación [19]

```python
def s_entropy(freq_list):
    freq_list = [element for element in freq_list if element != 0]
    sh_entropy = 0.0
    for freq in freq_list:
        sh_entropy += freq * np.log(freq)
    sh_entropy = -sh_entropy
    return(sh_entropy)
```

```python
def p_entropy(op):
    ordinal_pat = op
    max_entropy = np.log(len(ordinal_pat))
    p = np.divide(np.array(ordinal_pat), float(sum(ordinal_pat)))
    return(s_entropy(p)/max_entropy)
```

Crear una rutina que aplique sobre todos los archivos de la base de datos las rutina 1 y almacene los resultados en un dataframe donde se pueda discriminar nombre participante, condición (Parkinson, Huntington, ELA, Control) y los tres índices/características calculados por cada señal

```python
# Lista para almacenar los resultados de cada archivo
resultados = []

# Recorrer todos los archivos .ts
for archivo in archivos_ts:
    # Ruta completa del archivo
    ruta_archivo = extracted_data + "gait-in-neurodegenerative-disease-database-1.0.0/" + archivo

    # Procesar el archivo para obtener las señales filtradas
    señales_filtradas = procesar_archivo_ts(ruta_archivo)

    # Calcular entropía de permutación para cada señal
    ent_intervalo_izq = p_entropy(señales_filtradas[:, 0])  # Señal: Left Stride Interval
    ent_intervalo_der = p_entropy(señales_filtradas[:, 1])  # Señal: Right Stride Interval
    ent_balanceo_izq = p_entropy(señales_filtradas[:, 2])   # Señal: Left Swing Interval
    ent_balanceo_der = p_entropy(señales_filtradas[:, 3])   # Señal: Right Swing Interval

    # Calcular CV y SD para cada señal
    cv_intervalo_izq, sd_intervalo_izq = calcular_cv_sd(señales_filtradas[:, 0])
    cv_intervalo_der, sd_intervalo_der = calcular_cv_sd(señales_filtradas[:, 1])
    cv_balanceo_izq, sd_balanceo_izq = calcular_cv_sd(señales_filtradas[:, 2])
    cv_balanceo_der, sd_balanceo_der = calcular_cv_sd(señales_filtradas[:, 3])

    # Determinar el sujeto y la condición basada en el nombre del archivo
    sujeto = archivo.split('.')[0]

    if "control" in archivo.lower():
        estado = "Control"
    elif "park" in archivo.lower():
        estado = "Parkinson"
    elif "hunt" in archivo.lower():
        estado = "Huntington"
    elif "als" in archivo.lower():
        estado = "ALS"
    else:
        estado = "Desconocido"

    # Almacenar los resultados en un diccionario
    resultado = {
        'Sujeto': sujeto,
        'Estado': estado,
        'ent_intervalo_izq': ent_intervalo_izq,
        'ent_intervalo_der': ent_intervalo_der,
        'ent_balanceo_izq': ent_balanceo_izq,
        'ent_balanceo_der': ent_balanceo_der,
        'cv_intervalo_izq': cv_intervalo_izq,
        'cv_intervalo_der': cv_intervalo_der,
        'cv_balanceo_izq': cv_balanceo_izq,
        'cv_balanceo_der': cv_balanceo_der,
        'sd_intervalo_izq': sd_intervalo_izq,
        'sd_intervalo_der': sd_intervalo_der,
        'sd_balanceo_izq': sd_balanceo_izq,
        'sd_balanceo_der': sd_balanceo_der
    }
    resultados.append(resultado)

df_resultados = pd.DataFrame(resultados)
df_resultados.to_csv('resultados_marcha.csv', index=False)

df_resultados
```

#Comparación entre grupos:

Prueba de Shapiro-Wilk

**Hipótesis:**

p > alpha

*   H0: Los datos parecen normales (No se puede rechazar H0)
*   H1: Los datos no parecen normales (Se rechaza H0)

**Homocedasticidad:**

Prueba de Levene

**Hipótesis:**

*   H0: Los grupos tienen varianzas iguales (No se puede rechazar H0)
*   H1: Los grupos tienen varianzas diferentes (Se rechaza H0)

**Independencia**

Se asume que los grupos son independientes


```python
grupos = ['Control', 'ALS', 'Huntington', 'Parkinson']

# Definir las columnas que te interesa probar
columnas_a_probar = [
    'ent_intervalo_izq', 'ent_intervalo_der', 'ent_balanceo_izq', 'ent_balanceo_der',
    'cv_intervalo_izq', 'cv_intervalo_der', 'cv_balanceo_izq', 'cv_balanceo_der',
    'sd_intervalo_izq', 'sd_intervalo_der', 'sd_balanceo_izq', 'sd_balanceo_der'
]

# 1. Prueba de normalidad usando el test de Shapiro-Wilk
for columna in columnas_a_probar:
    print(f"\nResultados de la prueba de normalidad para {columna} por grupo:")
    for grupo in grupos:
        stat, p_value = stats.shapiro(df_resultados[df_resultados['Estado'] == grupo][columna])
        print(f"{grupo}: Estadístico W = {stat}, valor p = {p_value}")
        if p_value < 0.05:
            print(f"  -> La variable {columna} en el grupo {grupo} NO sigue una distribución normal (se rechaza H0).")
        else:
            print(f"  -> La variable {columna} en el grupo {grupo} sigue una distribución normal (no se rechaza H0).")

# 2. Prueba de homocedasticidad usando el test de Levene
for columna in columnas_a_probar:
    print(f"\nPrueba de Levene para homocedasticidad en {columna}:")
    valores_grupos = [df_resultados[df_resultados['Estado'] == grupo][columna] for grupo in grupos]
    stat, p_value = stats.levene(*valores_grupos)
    print(f"Estadístico W = {stat}, valor p = {p_value}")
    if p_value < 0.05:
        print(f"  -> Las varianzas de {columna} NO son iguales entre los grupos (se rechaza H0).")
    else:
        print(f"  -> Las varianzas de {columna} son iguales entre los grupos (no se rechaza H0).")
```

**Análisis de la Prueba de Normalidad:**

Estos resultados muestran que la distribución de las variables de intervalo y balanceo varían entre los grupos, y no siempre sigue una distribución normal. Esto es importante, ya que la normalidad de los datos es un supuesto importante para realizar análisis estadísticos paramétricos, como ANOVA o pruebas t. Debido a que varias de estas distribuciones no son normales, será necesario considerar pruebas no paramétricos (como la prueba de Kruskal-Wallis) para comparar los grupos de manera adecuada.

**Análisis de la Prueba de Homocedasticidad:**

Los resultados muestran que en ent_intervalo_izq, ent_balanceo_izq y cv_balanceo_izq las varianzas no son homogéneas entre los grupos, lo que podría complicar la realización de análisis paramétricos que asumen homocedasticidad, como ANOVA. Por otro lado en el resto de variables, no se rechaza la hipótesis de igualdad de varianzas, lo que implica que para estas variables se podrían aplicar análisis paramétricos siempre que se cumpla la normalidad o se utilicen transformaciones.



Dado que los resultados de las pruebas de normalidad muestran que en la mayoría de los casos las variables no siguen una distribución normal, y los resultados de la prueba de Levene indican que no se cumplen las condiciones de homocedasticidad en algunas de las variables, se decide aplicar la prueba no paramétrica de **Kruskal-Wallis**. Esta prueba es adecuada para comparar múltiples grupos cuando no se cumplen los supuestos de normalidad o igualdad de varianzas, como es el caso en este análisis. Al no requerir que los datos sigan una distribución normal, la prueba de Kruskal-Wallis permitirá identificar si existen diferencias estadísticamente significativas entre los grupos (Control, ALS, Huntington y Parkinson) para las variables de interés, de forma apropiada dadas las características de los datos.

**Prueba de Kruskal-Wallis:**

**Hipótesis:**


*   H0: Las distribuciones de los grupos son iguales, no hay diferencias significativas entre los grupos.
*   H1: Al menos un grupo proviene de una distribución diferente, es decir, existen diferencias significativas entre los grupos.



```python
# Definir los grupos y las columnas que te interesa probar
grupos = ['Control', 'ALS', 'Huntington', 'Parkinson']
columnas_a_probar = [
    'ent_intervalo_izq', 'ent_intervalo_der', 'ent_balanceo_izq', 'ent_balanceo_der',
    'cv_intervalo_izq', 'cv_intervalo_der', 'cv_balanceo_izq', 'cv_balanceo_der',
    'sd_intervalo_izq', 'sd_intervalo_der', 'sd_balanceo_izq', 'sd_balanceo_der'
]

# Prueba de Kruskal-Wallis
for columna in columnas_a_probar:
    print(f"\nPrueba de Kruskal-Wallis para {columna}:")
    valores_grupos = [df_resultados[df_resultados['Estado'] == grupo][columna] for grupo in grupos]
    stat, p_value = stats.kruskal(*valores_grupos)
    print(f"Estadístico H = {stat}, valor p = {p_value}")
    if p_value < 0.05:
        print(f"  -> Hay diferencias significativas entre los grupos para {columna} (se rechaza H0).")
    else:
        print(f"  -> No hay diferencias significativas entre los grupos para {columna} (no se rechaza H0).")
```

##**Análisis:**

En el análisis estadístico de las variables relacionadas con el balanceo postural, se utilizó la prueba no paramétrica de Kruskal-Wallis para evaluar la existencia de diferencias significativas entre los grupos de estudio: Control, ALS (Esclerosis Lateral Amiotrófica), Huntington y Parkinson. Las variables incluyeron la entropía, el coeficiente de variación, y la desviación estándar tanto en el pie izquierdo como en el pie derecho. Los resultados mostraron que en todos los casos, los valores p fueron extremadamente bajos (inferiores a 0.05), lo que indica diferencias estadísticamente significativas entre los grupos en todas las variables evaluadas. En particular, la entropía del balanceo y el coeficiente de variación mostraron diferencias entre los grupos para ambos pies, lo que sugiere que el balanceo postural varía entre personas sanas y aquellas con enfermedades neurodegenerativas como ALS, Huntington y Parkinson. Estos resultados tienen implicaciones importantes para el desarrollo de herramientas diagnósticas y terapéuticas basadas en el análisis del balanceo.

Este resultado es coherente con lo que se observó previamente en los análisis descriptivos y gráficos. A través de los boxplots y otras visualizaciones de estadística descriptiva, ya se evidenciaban diferencias en la dispersión y en la forma de los datos entre los grupos de estudio. En los gráficos, se podían observar patrones distintos en la distribución de las variables, con variaciones en la amplitud de los rangos intercuartílicos y la presencia de datos atípicos en algunos grupos, lo que demostraba que las características del balanceo postural no eran homogéneas entre los grupos. Además, las diferencias en la forma de las distribuciones dentro de los grupos reforzaban la hipótesis de que existen variaciones significativas en las variables de entropía y coeficiente de variación. Por tanto, los resultados obtenidos mediante la prueba de Kruskal-Wallis, que revelaron diferencias estadísticamente significativas entre los grupos, eran esperables y consistentes con los patrones observados previamente en el análisis descriptivo.
