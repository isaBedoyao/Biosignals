Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota mínima, la máxima, media y desviación típica.

```python
import pandas as pd

def serie(diccionario):
    notas = pd.Series(diccionario)

    minima = notas.min()
    maxima = notas.max()
    media = notas.mean()
    desviacion = notas.std()

    return {
        'nota mínima': minima,
        'nota máxima': maxima,
        'media': media,
        'desviación típica': desviacion
    }

diccionario = {
    'Daniel': 4.5,
    'Antonia': 3.8,
    'Sergio': 4.9,
    'Camila': 3.2,
    'Alexandra': 4.1
}

resultado = serie(diccionario)
print(resultado)
```
