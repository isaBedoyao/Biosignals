Escribir una funci√≥n que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota m√≠nima, la m√°xima, media y desviaci√≥n t√≠pica.

```python
import pandas as pd

def calcular_notas(notas_dict):
    # Convertir el diccionario a una serie de pandas
    notas = pd.Series(notas_dict)
    
    # Calcular las estad√≠sticas
    estadisticas = pd.Series({
        'Nota M√≠nima': notas.min(),
        'Nota M√°xima': notas.max(),
        'Media': notas.mean(),
        'Desviaci√≥n T√≠pica': notas.std()
    })
    
    return estadisticas

# Ejemplo de uso
notas_alumnos = {
    'Sergio': 4.5,
    'Isa': 3.8,
    'Dani': 4.3,
    'Dani2': 3.5,
    'Antonia': 4.9
}

resultado = calcular_notas(notas_alumnos)
print(resultado)
```
