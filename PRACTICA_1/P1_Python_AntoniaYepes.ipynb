Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota mínima, la máxima, media y desviación típica.

```python
import pandas as pd

def calcular_notas(notas_dict):
    # Convertir el diccionario a una serie de pandas
    notas = pd.Series(notas_dict)
    
    # Calcular las estadísticas
    estadisticas = pd.Series({
        'Nota Mínima': notas.min(),
        'Nota Máxima': notas.max(),
        'Media': notas.mean(),
        'Desviación Típica': notas.std()
    })
    
    return estadisticas

# Ejemplo de uso
notas_alumnos = {
    'Sergio': 4.5,
    'Isa': 3.8,
    'Dani': 4.3,
    'Dani2': 3.5,
    'Antonia': 4.9
}

resultado = calcular_notas(notas_alumnos)
print(resultado)
```
