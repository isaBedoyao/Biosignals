#Contextualización

La fibrilación auricular (AFIB) y la bradicardia sinusal (SB) son dos tipos de arritmias cardíacas que afectan el ritmo del corazón, pero lo hacen de maneras distintas. La **fibrilación auricular (AFIB)** es una arritmia común que se caracteriza por latidos rápidos e irregulares de las aurículas, lo que provoca una contracción auricular ineficiente. Este desorden eléctrico puede predisponer a la formación de coágulos y aumentar el riesgo de eventos graves como el accidente cerebrovascular. En un electrocardiograma (ECG), la AFIB se reconoce fácilmente por la ausencia de ondas P claras y la presencia de una actividad auricular rápida e irregular, acompañada de intervalos R-R igualmente irregulares [1].

Por otro lado, la **bradicardia sinusal (SB)** es un tipo de arritmia en la que el corazón late de manera más lenta de lo normal, generalmente con un ritmo regular. La SB se origina en el nodo sinusal, que es el marcapasos natural del corazón. Aunque puede considerarse normal en personas entrenadas o en reposo profundo, en algunos casos puede estar relacionada con enfermedades o problemas del sistema de conducción eléctrica del corazón. En el ECG, la SB se manifiesta con ondas P normales pero con intervalos R-R prolongados debido a la baja frecuencia cardíaca [2].

El propósito del proyecto es encontrar factores diferenciadores en los registros de ECG que permitan distinguir entre la fibrilación auricular y la bradicardia sinusal. Si bien ambas arritmias afectan el ritmo cardíaco, las características eléctricas que se observan en el ECG son diferentes, y este análisis busca mejorar la precisión diagnóstica al resaltar esos patrones específicos. Detectar estas diferencias en los ECG permitirá un diagnóstico más eficiente y un manejo clínico adecuado para cada condición.

### Formula 1 - Detrend

````python
def apply_detrend(df, column_name='II_Derivada'):

     # Convertir la columna en listas de floats si es necesario
    df[column_name] = df[column_name].apply(
        lambda x: np.array([float(i) for i in x.split(',')]) if isinstance(x, str) else x
    )

    # Aplicar detrend a cada fila
    df[f'Detrended {column_name}'] = df[column_name].apply(lambda x: detrend(x))

    return df
````

### Formula 2 - Compression

````python
def compress_signal(signal):
    x_min = np.min(signal)  # Valor mínimo de la señal
    x_max = np.max(signal)  # Valor máximo de la señal
    compressed_signal = (signal - x_min) / (x_max - x_min)  # Fórmula de compresión
    return compressed_signal
````

````python
def apply_compression(df, column_name='Detrended II_Derivada'):
    # Convertir la columna en listas de floats si es necesario
    df[column_name] = df[column_name].apply(
        lambda x: np.array([float(i) for i in x.split(',')]) if isinstance(x, str) else x
    )
    # Aplicar la compresión a cada fila de la columna especificada
    df[f'Compressed {column_name}'] = df[column_name].apply(lambda x: compress_signal(x))

    return df
````

### Formula 3 - Biespectro

````python
# Función para calcular el bispectro de una señal
def calculate_bispectrum(signal, segment_length, fs):
    num_segments = len(signal) // segment_length
    max_freq = 20  # Frecuencia máxima a analizar en Hz
    max_freq_idx = int(max_freq * segment_length / fs)
    bispectrum_sum = np.zeros((max_freq_idx, max_freq_idx), dtype=complex)

    # Calcular la FFT para cada segmento
    for segment_index in range(num_segments):
        segment_data = signal[segment_index * segment_length:(segment_index + 1) * segment_length]
        dft_segment = fft(segment_data)

        # Calcular el bispectro solo para la región triangular
        for f1 in range(max_freq_idx):
            for f2 in range(max_freq_idx - f1):  # Asegura que f1 + f2 <= fmax
                bispectrum_sum[f1, f2] += dft_segment[f1] * dft_segment[f2] * np.conj(dft_segment[f1 + f2])

    # Devolver la suma de los productos (bispectro)
    return bispectrum_sum / num_segments
````

### Formula 4 - Bicoherencia

````python
# Función para calcular la bicoherencia a partir del bispectro
def calculate_bicoherence(signal, segment_length, fs):
    num_segments = len(signal) // segment_length
    max_freq = 20  # Frecuencia máxima a analizar en Hz
    max_freq_idx = int(max_freq * segment_length / fs)
    bispectrum_sum = np.zeros((max_freq_idx, max_freq_idx), dtype=complex)
    abs_sum = np.zeros((max_freq_idx, max_freq_idx))

    # Calcular la FFT para cada segmento y acumular valores para el numerador y el denominador
    for segment_index in range(num_segments):
        segment_data = signal[segment_index * segment_length:(segment_index + 1) * segment_length]
        dft_segment = fft(segment_data)

        # Calcular el bispectro y acumular el numerador y el denominador
        for f1 in range(max_freq_idx):
            for f2 in range(f1, max_freq_idx - f1):  # Asegura que f1 + f2 <= fmax
                value = dft_segment[f1] * dft_segment[f2] * np.conj(dft_segment[f1 + f2])
                bispectrum_sum[f1, f2] += value
                abs_sum[f1, f2] += np.abs(value)

    # Calcular la bicoherencia
    bicoherence = np.zeros((max_freq_idx, max_freq_idx))
    non_zero_denominator = abs_sum > 0
    bicoherence[non_zero_denominator] = np.abs(bispectrum_sum[non_zero_denominator]) / abs_sum[non_zero_denominator]

    return bicoherence
````

````python
# Función para obtener el promedio de varias señales
def average_signals(df, column_name, num_signals):
    # Obtener las primeras 'num_signals' señales de la columna especificada
    signals = df[column_name].iloc[:num_signals].values

    # Convertir las señales a una matriz para facilitar el promedio
    signals_array = np.array(signals.tolist())

    # Calcular el promedio a lo largo del eje 0 (promedio punto a punto)
    average_signal = np.mean(signals_array, axis=0)

    return average_signal
````

El **biespectro** y la **bicoherencia** son herramientas matemáticas utilizadas en el análisis de señales no lineales, específicamente en el contexto de análisis de señales fisiológicas, como las señales de ECG.

El **biespectro** se deriva del concepto de espectro de Fourier, pero amplía su análisis a la detección de relaciones entre múltiples componentes de frecuencia. Mientras que el análisis espectral tradicional examina la energía en diferentes frecuencias de una señal, el biespectro añade una capa adicional al identificar correlaciones no lineales entre diferentes frecuencias dentro de la señal, es decir, el biespectro examina la interacción entre pares de componentes frecuenciales (f1, f2) y un tercer componente en la suma de esas frecuencias (f1 + f2), lo que permite estudiar el acoplamiento de fase cuadrática (QPC). De esta forma, es posible investigar las interacciones entre tres frecuencias, lo que es útil para detectar estructuras no lineales en la señal, como la presencia de armónicos o modulaciones.

$$
b(f_1,f_2) = \sum_{j=1}^{N} X_j(f_1) X_j(f_2) X_j^*(f_1 + f_2)
$$


Por otro lado, la **bicoherencia** es una función derivada del biespectro que se utiliza para normalizar las interacciones entre esas frecuencias, permitiendo identificar qué tan fuerte es la relación entre ellas. Esta se trata de una versión normalizada del biespectro, que es útil para eliminar influencias no deseadas de la energía de la señal, facilitando la detección de fases acopladas entre componentes frecuenciales. La bicoherencia proporciona una medida que varía entre 0 y 1, donde los valores cercanos a 1 indican una fuerte correlación no lineal entre tres frecuencias de la señal, mientras que los valores cercanos a 0 sugieren una relación débil o inexistente.  En este contexto, la bicoherencia es útil para discernir si las correlaciones observadas en el biespectro son significativas o si se deben al ruido.

$$
B(f_1,f_2) = \frac{\left|\sum_{j=1}^{N} X_j(f_1) X_j(f_2) X_j^*(f_1 + f_2)\right|}{\sum_{j=1}^{N} \left| X_j(f_1) X_j(f_2) X_j^*(f_1 + f_2) \right|}
$$


Ambos métodos son útiles cuando se trabaja con sistemas complejos y no lineales, como los procesos fisiológicos que gobiernan el sistema cardiovascular. En el caso de señales de HRV, estos análisis permiten identificar interacciones no lineales que no pueden ser detectadas con el análisis espectral tradicional, ayudando a revelar mecanismos subyacentes de control que influyen en la frecuencia cardíaca [3].

$$ s(t_k) = \frac{c(t_k) - x_{\text{min}}}{x_{\text{max}} - x_{\text{min}}} $$

# 4. Para cada señal extraer la frecuencia que contiene la máxima potencia y Gráfiquela

````python
def max_frequency_power(signal, fs=500):
    # Realizar la FFT
    EKG = fft(signal)
    N = len(signal)  # Número de puntos en la señal

    # Calcular las frecuencias correspondientes
    f = np.arange(0, fs, fs/N)

    # Calcular el espectro de potencia (PSD)
    power_spectrum = np.power(np.abs(EKG), 2) / N

    # Extraer la frecuencia que contiene la máxima potencia
    max_power_index = np.argmax(power_spectrum[:N//2])  # Solo tomamos la mitad positiva
    max_frequency = f[max_power_index]

    return max_frequency, f[:N//2], power_spectrum[:N//2]  # Devolver el espectro de potencia y las frecuencias
````

# 5. Para cada señal sumar el número de frecuencias que son significativas para bicoherencia, cuales tienen un valor superior a:

 $$ \sqrt{\frac{9.2}{2 \cdot N}} $$

````python
def calculate_significant_bicoherence(bicoherence, num_segments):
    # Calcular el umbral significativo para bicoherencia
    threshold = np.sqrt(9.2 / (2 * num_segments))

    # Contar las frecuencias significativas que superan el umbral
    significant_bicoherence = bicoherence > threshold

    # Sumar el número de frecuencias significativas
    sFSB = np.sum(significant_bicoherence)

    return sFSB, threshold
````

````python
# Función para sumar el número de frecuencias significativas para cada señal
def count_significant_frequencies(df, column_name, segment_length, fs):
    # Listas para almacenar los resultados
    significant_counts = []
    thresholds = []

    # Iterar sobre cada señal en la columna
    for signal in df[column_name]:
        # Calcular la bicoherencia para la señal
        bicoherence = calculate_bicoherence(signal, segment_length, fs)

        # Número de segmentos
        num_segments = len(signal) // segment_length

        # Calcular las frecuencias significativas utilizando la función proporcionada
        sFSB, threshold = calculate_significant_bicoherence(bicoherence, num_segments)

        # Almacenar el número de frecuencias significativas y el umbral
        significant_counts.append(sFSB)
        thresholds.append(threshold)

    return significant_counts, thresholds
````

### Numero de frecuencias significativas - SB

````python
# Parámetros
segment_length = 500  # Longitud del segmento
fs = 500  # Frecuencia de muestreo

# Aplicar la función para contar frecuencias significativas en la columna 'Detrended Compressed Detrended II_Derivada'
significant_counts_sb, thresholds = count_significant_frequencies(df_II_SB, 'Detrended Compressed Detrended II_Derivada', segment_length, fs)
````

# 6. Crear una rutina que aplique sobre todos los archivos de la base de datos las rutina 3 al 5 y almacene los resultados en un dataframe donde se pueda registro, tipo de patología y el frecuencia de máxima potencia (fMP) y suma de frecuencias significativas en bicoherencia (sFSB):

````python
def process_signals(df, estado, significant_counts):
    result_rows = []

    for idx, row in df.iterrows():
        registro = row['FileName']  # Nombre del registro
        signal = row['Detrended Compressed Detrended II_Derivada']  # Tomar la columna correcta

        # Asegurarse de que la señal es un array de floats
        if isinstance(signal, list):
            signal = np.array(signal)

        # Calcular la frecuencia de máxima potencia (fMP)
        fmp, _, _ = max_frequency_power(signal, fs=500)

        # Obtener el número de frecuencias significativas (sFSB) de los valores ya calculados
        sfsb = significant_counts[idx]

        # Guardar los resultados en una fila
        result_rows.append({
            'Registro': registro,
            'Estado': estado,
            'fMP': fmp,
            'sFSB': sfsb
        })

    return pd.DataFrame(result_rows)
````

**Hipótesis:**

p > alpha

*   H1: Los datos no distribuyen normal (Se rechaza H0)

**Homocedasticidad:**

Prueba de Levene

**Hipótesis:**

*   H0: Los grupos tienen varianzas iguales (No se puede rechazar H0)
*   H1: Los grupos tienen varianzas diferentes (Se rechaza H0)

**Independencia**

Se asume que los grupos son independientes

````python
import scipy.stats as stats

# Nivel de significancia (alpha)
alpha = 0.05

# Filtrar los datos por grupos SB y AFIB
sb_data_fmp = df_final[df_final['Estado'] == 'SB']['fMP']
afib_data_fmp = df_final[df_final['Estado'] == 'AFIB']['fMP']

sb_data_sfsb = df_final[df_final['Estado'] == 'SB']['sFSB']
afib_data_sfsb = df_final[df_final['Estado'] == 'AFIB']['sFSB']

# Prueba de normalidad Kolmogorov-Smirnov para fMP (SB y AFIB)
ks_stat_sb_fmp, p_value_sb_fmp = stats.kstest(sb_data_fmp, 'norm')
ks_stat_afib_fmp, p_value_afib_fmp = stats.kstest(afib_data_fmp, 'norm')

# Prueba de normalidad Kolmogorov-Smirnov para sFSB (SB y AFIB)
ks_stat_sb_sfsb, p_value_sb_sfsb = stats.kstest(sb_data_sfsb, 'norm')
ks_stat_afib_sfsb, p_value_afib_sfsb = stats.kstest(afib_data_sfsb, 'norm')

# Función para interpretar los resultados según las hipótesis del usuario
def interpretar_resultado(p_value, variable, estado):
    if p_value > alpha:
        print(f"No se puede rechazar H0 para {variable} en {estado}: Los datos distribuyen normal (p-valor = {p_value:.4f})")
    else:
        print(f"Se rechaza H0 para {variable} en {estado}: Los datos no distribuyen normal (p-valor = {p_value:.4f})")

# Interpretación de los resultados para fMP
interpretar_resultado(p_value_sb_fmp, 'fMP', 'SB')
interpretar_resultado(p_value_afib_fmp, 'fMP', 'AFIB')

# Interpretación de los resultados para sFSB
interpretar_resultado(p_value_sb_sfsb, 'sFSB', 'SB')
interpretar_resultado(p_value_afib_sfsb, 'sFSB', 'AFIB')

````

 Se realizó la prueba de Shapiro-Wilk en las características de la Frecuencia de Máxima Potencia (fMP) y la Suma de Frecuencias Significativas en Bicoherencia (sFSB) de los dos grupos (BS y AFIB). Los resultados mostraron que en todos los casos el p-valor fue 0.0000, lo que indica que las distribuciones de estas variables no siguen una distribución normal. La falta de normalidad en los datos hace necesario el uso de pruebas paramétricas convencionales, como la prueba t de Student o ANOVA, las cuales asumen que los datos deben distribuirse normalmente. Por lo tanto, se optó por aplicar pruebas no paramétricas.

````python
# Prueba de Levene para fMP
levene_stat_fmp, p_value_fmp = stats.levene(sb_data_fmp, afib_data_fmp)
print(f"Resultado de la prueba de Levene para fMP:")
print(f"Estadístico = {levene_stat_fmp}, p-valor = {p_value_fmp}")
if p_value_fmp > alpha:
    print(f"No se puede rechazar H0: Las varianzas de fMP son iguales (p-valor = {p_value_fmp:.4f})")
else:
    print(f"Se rechaza H0: Las varianzas de fMP no son iguales (p-valor = {p_value_fmp:.4f})")

# Prueba de Levene para sFSB
levene_stat_sfsb, p_value_sfsb = stats.levene(sb_data_sfsb, afib_data_sfsb)
print(f"\nResultado de la prueba de Levene para sFSB:")
print(f"Estadístico = {levene_stat_sfsb}, p-valor = {p_value_sfsb}")
if p_value_sfsb > alpha:
    print(f"No se puede rechazar H0: Las varianzas de sFSB son iguales (p-valor = {p_value_sfsb:.4f})")
else:
    print(f"Se rechaza H0: Las varianzas de sFSB no son iguales (p-valor = {p_value_sfsb:.4f})")
````

 Se utilizó la prueba de Levene para determinar si las varianzas de fMP y sFSB eran iguales entre los grupos de BS y AFIB. Los resultados mostraron que las varianzas en ambas características no son iguales, ya que los p-valores fueron 0.0000 en todos los casos. Esto indica una falta de homocedasticidad, lo cual refuerza la necesidad de utilizar pruebas no paramétricas, dado que muchas pruebas paramétricas también requieren la igualdad de varianza entre los grupos como supuesto fundamental.

````python
# Prueba de Mann-Whitney para fMP
mw_stat_fmp, p_value_fmp = stats.mannwhitneyu(sb_data_fmp, afib_data_fmp, alternative='two-sided')
print(f"Resultado de la prueba U de Mann-Whitney para fMP:")
print(f"Estadístico = {mw_stat_fmp}, p-valor = {p_value_fmp}")
if p_value_fmp > alpha:
    print(f"No se puede rechazar H0: No hay diferencias en la tendencia central de fMP entre SB y AFIB (p-valor = {p_value_fmp:.4f})")
else:
    print(f"Se rechaza H0: Hay diferencias en la tendencia central de fMP entre SB y AFIB (p-valor = {p_value_fmp:.4f})")

# Prueba de Mann-Whitney para sFSB
mw_stat_sfsb, p_value_sfsb = stats.mannwhitneyu(sb_data_sfsb, afib_data_sfsb, alternative='two-sided')
print(f"\nResultado de la prueba U de Mann-Whitney para sFSB:")
print(f"Estadístico = {mw_stat_sfsb}, p-valor = {p_value_sfsb}")
if p_value_sfsb > alpha:
    print(f"No se puede rechazar H0: No hay diferencias en la tendencia central de sFSB entre SB y AFIB (p-valor = {p_value_sfsb:.4f})")
else:
    print(f"Se rechaza H0: Hay diferencias en la tendencia central de sFSB entre SB y AFIB (p-valor = {p_value_sfsb:.4f})")
````

Al comparar la Frecuencia de Máxima Potencia (fMP) y la Suma de Frecuencias Significativas en Bicoherencia (sFSB) entre las señales de Bradicardia Sinusal (SB) y Fibrilación Auricular (AFIB), se observa que hay diferencias significativas en la tendencia central de ambas métricas entre los dos grupos.

La fMP representa la frecuencia dominante en el espectro de la señal. En el análisis de fMP, la prueba de Mann-Whitney U muestra que hay diferencias significativas entre SB y AFIB, con un p-valor extremadamente bajo (2.867e-118). Esto indica que la tendencia central de fMP varía notablemente entre los dos grupos. Al revisar el boxplot de fMP, se observa que la dispersión es más amplia en el grupo de AFIB, lo que sugiere que la fibrilación auricular produce una mayor variabilidad en las frecuencias dominantes, confirmando los análisis previos. Esto concuerda con la naturaleza más irregular de AFIB, que genera componentes de frecuencia más diversas en comparación con la bradicardia sinusal, que es más ordenada y regular.

La sFSB mide el acoplamiento de fase entre las frecuencias significativas de la señal, lo cual proporciona una idea de cómo interactúan las distintas frecuencias en la señal de ECG. La prueba de Mann-Whitney U también revela una diferencia significativa en la sFSB entre SB y AFIB, con un p-valor incluso más bajo (1.528e-274). El boxplot de sFSB muestra que los valores de esta métrica son más altos en SB, lo que sugiere un mayor acoplamiento de fase en la bradicardia sinusal. Esto es consistente con la naturaleza más sincronizada de SB, donde las frecuencias tienden a estar más organizadas y acopladas en comparación con AFIB, que presenta interacciones de fase más desordenadas debido a su comportamiento eléctrico irregular.
